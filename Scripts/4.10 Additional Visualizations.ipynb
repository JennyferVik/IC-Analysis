### a few additional visualizations for IC

### Importing libraries and data

```python
# libraries
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
```

```python
# data
path = r'/Users/jennyfevik/Desktop/IC Basket Analysis'
```

```python
ic_data = pd.read_pickle(os.path.join(path, 'Data', 'Prepared', 'ic_data_final.pkl'))
```

```python
# grouping region and user spending
region_spending_counts = ic_data.groupby(['Region','user_spending']).size().unstack(fill_value=0)
```

```python
bar_region_spending = region_spending_counts.plot.bar(stacked=True)
plt.xlabel('Region', fontsize = 10,)
plt.ylabel('Number of Items Orderd', fontsize = 10)
plt.title('User Spending By Region', fontsize = 10)
plt.tight_layout()
plt.show()
```

```python
# exporting chart
bar_region_spending.figure.savefig(os.path.join(path, 'Analysis', 'Visualizations', 'bar_region_spending.png'))
```

```python
# group by frequency and loyalty
loyalty_frequency_counts = ic_data.groupby(['loyalty_flag','user_frequency']).size().unstack(fill_value=0)
```

```python
bar_loyatly_frequency = loyalty_frequency_counts.plot.bar(stacked=True)
plt.xlabel('Loyalty Group', fontsize = 10,)
plt.ylabel('Number of Items Orderd', fontsize = 10)
plt.title('Orders by User Frequency', fontsize = 10)
plt.tight_layout()
plt.show()
```

```python
# export frequency and loyalty
bar_loyatly_frequency.figure.savefig(os.path.join(path, 'Analysis', 'Visualizations', 'bar_loyalty_frequency.png'))
```

```python
#Pie chart of percentages of customer type groups
colour = ['blue', 'green', 'orange']
custpie = ic_data['loyalty_flag'].value_counts().plot.pie(figsize = (7,8), colors = colour, autopct = '%1.0f%%')
plt.ylabel('')
plt.title('Type of Customers', fontsize=13)
plt.show()
```

```python
# export loyalty pie
custpie.figure.savefig(os.path.join(path, 'Analysis', 'Visualizations', 'cust_loyalty.png'))
```

```python
# loyalty and spending
loyalty_spend_counts = ic_data.groupby(['loyalty_flag','user_spending']).size().unstack(fill_value=0)
```

```python
bar_loyal_spend = loyalty_spend_counts.plot.bar(stacked=True)
plt.xlabel('Spending Group', fontsize = 10,)
plt.ylabel('Number of Items Orderd)', fontsize = 10)
plt.title('Orders by User Spending', fontsize = 10)
plt.tight_layout()
plt.show()
```

```python
# export bar
bar_loyal_spend.figure.savefig(os.path.join(path, 'Analysis', 'Visualizations', 'bar_loyal_spend.png'))
```

```python
# group region and loyalty
region_loyal_counts = ic_data.groupby(['Region','loyalty_flag']).size().unstack(fill_value=0)
```

```python
bar_region_loyal = region_loyal_counts.plot.bar(stacked=True)
plt.xlabel('Region', fontsize = 10,)
plt.ylabel('Number of Items Orderd (in millions)', fontsize = 10)
plt.title('Loyalty by Region', fontsize = 10)
plt.tight_layout()
plt.show()
```

```python
# exporting
bar_region_loyal.figure.savefig(os.path.join(path, 'Analysis', 'Visualizations', 'bar_region_loyal.png'))
```
